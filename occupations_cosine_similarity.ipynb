```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_distances
```

```python
occupations = pd.read_csv("occupations.csv")
```

```python
similarity_matrix = cosine_distances(occupations.iloc[:, 1:])
```

```python
test1 = [None for x in range(0, len(similarity_matrix))]
for x in range(0, len(similarity_matrix)):
    closestCity = None
    for y in range(0, len(similarity_matrix[x])):
        matrix_data = (occupations['CBSA'].iloc[x],
             occupations['CBSA'].iloc[y],
             float(similarity_matrix[x][y]) * 1000)
        if (not closestCity or matrix_data[2] < closestCity[1]) and matrix_data[0] != matrix_data[1]:
            closestCity = (matrix_data[1], matrix_data[2])
    print(occupations['CBSA'].iloc[x], closestCity)
    test1[x] = closestCity[1]
```
